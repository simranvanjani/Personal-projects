#preprocessing 



*   ID number and Class column dropped from df to be fed to kmeans

*   Feature scaling using standard scaler 



```python
df1 = df.drop('id number', axis = 1)
```

```python
df1.head()
```

```python
y = df1['Class']
x = df1.drop('Class', axis=1)
```

```python
x.head()
```

```python
cols = x.columns
from sklearn.preprocessing import StandardScaler
s_c=StandardScaler()
df2 =s_c.fit_transform(x)
df2 = pd.DataFrame(df2,columns=['Clump_Thickness', 'Uniformity_of_cell_size',
       'Uniformity_of_cell_shape', 'Marginal_Adhesion',
       'Single_Epithelial_Cell_Size', 'Bare_Nuclei', 'Bland_Chromatin',
       'Normal_Nucleoli', 'Mitoses'])
df2.head(10)
```

#model


*   Kmeans initialized using n = 2
*   df fit to kmeans
*   clusters centres and inertia calculated 
*   Predicted class converted to convention mentioned in original data
*   A new dataframe formed containing ID number, class and predicted class



```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2,n_init=10)
```

```python
kmeans.fit(df2)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
pred_labels=pd.DataFrame(kmeans.labels_)
```

```python
pred_labels.columns=['Class']
```

```python
pred_labels
```

```python
def converter(class_):
    if class_==0:
        return 2
    else:
        return 4
a= pred_labels['Class'].apply(converter)
a
```

```python
a_df =pd.DataFrame(a)
```

```python
pred =df[['id number', 'Class']].copy()
pred['Predicted_Cluster']=pd.DataFrame(a)
pred
```

#Calcualting Presision, Recall, F1-Score of actual and Kmeans predicted labels using Confusion Matrix and Classification report. Also finding Silhouette score

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y,a_df))
print(classification_report(y,a_df))
```

#ErrorRate calculation by using user defined function as given in the documentation

```python
def error_func(actual,predicted):
    x=pred[(pred.Class==actual)&(pred.Predicted_Cluster==predicted)]
    y=pred[(pred.Predicted_Cluster==actual)]
    return float(len(x)/len(y))
```

```python
error_B=error_func(2,4)
print(error_B)
error_M=error_func(4,2)
print(error_M)
```
