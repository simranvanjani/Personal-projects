```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from matplotlib import pyplot
%matplotlib inline
```

#Import Data Set by going in file in LHS 
df= pd.read_csv("cab_rides.csv")

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
df1 = df.drop(['time_stamp','destination', 'source', 'id', 'product_id'], axis = 1) #removing unecessary columns 
df1.head()
```

```python
df_dum= pd.get_dummies(data = df1, prefix=["cab_type","name"]).head()
print(df_dum.columns.values)
df_dum.head()
```

```python
X = df_dum.drop('price',axis=1) # Independent variable
y = df_dum['price'] # dependent variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=60)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
trainSet = pd.concat([X_train, y_train], axis = 1)
trainSet.head()
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
predVal = pd.concat([X_test, y_test], axis = 1)
predVal.head()
```

```python
print("Training set accuracy = " + str(regressor.score(X_train, y_train)))
print("Test set accuracy = " + str(regressor.score(X_test, y_test)))
```
